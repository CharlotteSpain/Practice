IF STATEMENTS 

Statement -> condition (if the condition is true, then we do the action - otherise we just skip it):

I wake up
If I'm hungry
    I eat breakfast 

Creating an 'If statement'

```python
is_male = True

if is_male:
    print("you are a male")
```

```python
is_male = False

if is_male:
    print("you are a male")
else:
    print("you are not a male")
```

```python
is_male = True
is_tall = True

if is_male or is_tall:
    print("you are a male or tall or both")
else:
    print("you are neither a male or tall")
```

```python
is_male = True
is_tall = False

if is_male or is_tall:
    print("you are a male or tall or both")
else:
    print("you are neither a male or tall")
```

```python
is_male = False
is_tall = False

if is_male or is_tall:
    print("you are a male or tall or both")
else:
    print("you are neither a male or tall")
```

The 'and' + 'or' conditions work 

elif = else if

```python
is_male = False
is_tall = True

if is_male and is_tall:
    print("you are a male and tall or both")
else:
    print("you are either a not male or not tall")
```

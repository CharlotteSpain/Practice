a parameter is  a piece of information that we give the function

```python
def say_hi(name):
    print("Hello "+ name)
    
say_hi("mike")
say_hi("steve")
```

```python
def say_hi(name, age):
    print("Hello "+ name + ", you are " + age)
    
say_hi("mike", "15")
say_hi("steve", "36")
```

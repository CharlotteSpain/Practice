IF STATEMENTS AND COMPARISONS

This code is able to tell us which number is the biggest out of number 1, number 2, and number 3. It returns to us the largest number.

```python
def max_num(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3
print(max_num(300,40,5))
```
